```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos

# center of the disk, O
omega = 1.0
r = 2.0
v = omega * r
x = lambda t: v * t
y = lambda t: r

a = 0.3
p = lambda t: np.array([x(t) + cos(-omega * t) * a, y(t) + sin(-omega * t) * a])

pts = np.array([p(t) for t in np.linspace(0, 20, 500)])
plt.plot(pts[:, 0], pts[:, 1])
plt.show()
```
